```python
import pandas as pd
import numpy as np
import pylab
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Burger Dam 050 450 05 - Burger Dam 050 450 05.csv', names=('Potential', 'Current', 'ln(I)', 'Potential_Difference', 'lnI', 'Log_Current'))
```

```python
print(df)
```

```python
data = np.asarray(df)
data.shape

Potential = data[:,0]
Current = data[:,1]
lnI = data[:,4]
Log_Current = data[:,5]

print(data)
```

```python
plt.style.use(['seaborn-pastel'])
plt.title('Current and Potential')
plt.ylabel('Current (mA)')
plt.xlabel('Potential (V)')
pylab.xlim(-400, -50)
pylab.ylim(-1.4E-5, 1.1E-3)
plt.grid()

plt.plot(df.Potential, df.Current)
plt.savefig('./CurrentvPotential.svg')
```

```python
#Finding the saturated ion current
#Saturated ion currrent is the y-intercept

from statistics import mean

xI_is = np.array(Potential[:177])
yI_is = np.array(Current[:177])

def best_fit_slope(xI_is,yI_is):
    mI_is = (((mean(xI_is)*mean(yI_is)) - mean(xI_is*yI_is)) /
         ((mean(xI_is)*mean(xI_is)) - mean(xI_is*xI_is)))
    return mI_is

mI_is = best_fit_slope(xI_is,yI_is)

def best_fit_slope_and_intercept(xI_is,yI_is):
    mI_is = (((mean(xI_is)*mean(yI_is)) - mean(xI_is*yI_is)) /
         ((mean(xI_is)*mean(xI_is)) - mean(xI_is*xI_is)))
    
    bI_is = mean(yI_is) - mI_is*mean(xI_is)
    return mI_is,bI_is

mI_is,bI_is = best_fit_slope_and_intercept(xI_is,yI_is)
    
print(mI_is,bI_is)

Regression_I_is = [mI_is * x + bI_is for x in xI_is]
```

```python
#Plotting linear fit for the ion saturation region

plt.style.use(['seaborn-pastel'])
plt.title('Ion Saturation')
plt.ylabel('Current (mA)')
plt.xlabel('Potential (V)')
pylab.xlim(-400, -50)
pylab.ylim(-1.4E-5, 1.1E-3)
plt.grid()

plt.plot(xI_is, Regression_I_is, '-', color='black')
plt.plot(df.Potential, df.Current)
plt.savefig('./IonSaturation.svg')
```

```python
#Finding saturated electron current
#Saturated electron current is y-intercept

xI_es = np.array(Potential[335:])
yI_es = np.array(Current[335:])


def best_fit_slope(xI_es,yI_es):
    mI_es = (((mean(xI_es)*mean(yI_es)) - mean(xI_es*yI_es)) /
         ((mean(xI_es)*mean(xI_es)) - mean(xI_es*xI_es)))
    return mI_es

mI_es = best_fit_slope(xI_es,yI_es)

def best_fit_slope_and_intercept(xI_es,yI_es):
    mI_es = (((mean(xI_es)*mean(yI_es)) - mean(xI_es*yI_es)) /
         ((mean(xI_es)*mean(xI_es)) - mean(xI_es*xI_es)))
    
    bI_es = mean(yI_es) - mI_es*mean(xI_es)
    return mI_es,bI_es

mI_es,bI_es = best_fit_slope_and_intercept(xI_es,yI_es)
    
print(mI_es,bI_es)

Regression_I_es = [mI_es * x + bI_es for x in xI_es]
```

```python
#Plotting linear fit for the electron saturation region

plt.style.use(['seaborn-pastel'])
plt.title('Electron Saturation')
plt.ylabel('Current (mA)')
plt.xlabel('Potential (V)')
pylab.xlim(-400, -50)
pylab.ylim(-2.0E-5, 1.1E-3)
plt.grid()

plt.plot(xI_es, Regression_I_es, '-', color='black')
plt.plot(df.Potential, df.Current)
plt.savefig('./ElectronSaturation.svg')
```
